## Achieving perfect communication over an imperfect communication channel
One canonical example is the **binary symmetric channel**. Consider a noisy disk drive that transmits each bit correctly with
probability (1−f) and incorrectly with probability f.

In order to reduce communication error, we accept a noisy channel, and add communication systems to detect and correct the errors introduced by the channel.

$$(s) → Encoder → (t) → Noisy Channel → (r) → Decoder → (ŝ)$$

```python
from matplotlib import pyplot as plt
import numpy as np
from PIL import Image
```

```python
def image2array(filename):
    img = Image.open(filename)
    ary = np.array(img)
    r,g,b = np.split(ary,3,axis=2)
    return r>0 # return red channel for now
```

```python
def array2plot(array):
    return plt.imshow(array, cmap=plt.get_cmap('gray'))
```

```python
a = image2array("./img/happyman.jpg")
array2plot(a)
```
